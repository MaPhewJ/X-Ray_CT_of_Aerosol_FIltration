## Rendering of an ash particle sliced so that internal pores are visible 

```python
vol = io.volread(f'F:/Ptycho/336771_TIFF_delta_FBP_ram-lak_freqscl_1.00/336771_FBP_smooth_matched.tiff')[680:750,980:1185,620:865]
vol = Volume(vol)
lego = vol.legosurface(140,255).add_scalarbar()
lego.cmap('afmhot', on='cells', vmin=50, vmax=240).add_scalarbar()
show(lego, axes=True, backend='ipyvtk')
```
